```python
import pandas as pd
import numpy as np

import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv('/content/train (2).csv',usecols=['Age','Fare','Survived'])
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True) #REMOVING NULL VALUES WITH MEAN OF AGE
```

```python
df.isnull().sum()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
clf=LogisticRegression()
clf1=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
clf1.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
y_pred1=clf1.predict(X_test)
```

```python
print("accuracy score for LR=",accuracy_score(y_test,y_pred))
print("accuracy score for DT=",accuracy_score(y_test,y_pred1))
```

NOW WE APPLY LOG TRANSFORM ON THEM

```python
trf=FunctionTransformer(func=np.sqrt)#func=np.log1p=for log sqrt for square root
```

```python
X_train_transformed=trf.fit_transform(X_train)
X_test_transformed=trf.transform(X_test)
```

```python
clf.fit(X_train_transformed,y_train)
clf1.fit(X_train_transformed,y_train)
```

```python
y_preddd=clf.predict(X_test_transformed)
y_preddd1=clf1.predict(X_test_transformed)
```

```python
print("accuracy score for LR=",accuracy_score(y_test,y_preddd))
print("accuracy score for DT=",accuracy_score(y_test,y_preddd1))
```

Using column transformer because we dont need to apply function transformation on age


```python
clff=LogisticRegression()
clff1=DecisionTreeClassifier()
```

```python
hello=ColumnTransformer([
    ('log',FunctionTransformer(func=np.log1p),['Fare'])
],remainder='passthrough')
```

```python
X_train_transformedd=hello.fit_transform(X_train)
X_test_transformedd=hello.transform(X_test)
```

```python
X_train_transformedd
```

```python
clff.fit(X_train_transformedd,y_train)
clff1.fit(X_train_transformedd,y_train)
```

```python
yy=clf.predict(X_test_transformedd)
yyy=clf1.predict(X_test_transformedd)
```

```python
print("accuracy score for LR=",accuracy_score(y_test,yy))
print("accuracy score for DT=",accuracy_score(y_test,yyy))
```
