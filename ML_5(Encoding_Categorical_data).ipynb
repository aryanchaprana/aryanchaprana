### ***ORDINAL ENCODING***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/customer.csv')
```

```python
df.sample(6)
```

```python
df=df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X=df.iloc[:, :-1]
y=df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
X_train=oe.transform(X_train)
X_test=oe.transform(X_test)
```

### ***LABEL ENCODER***

```python
from sklearn.preprocessing import LabelEncoder #We should use label encoding only on the output data not the input data
le=LabelEncoder()
le.fit(y_train)
y_train=le.transform(y_train)
y_test=le.transform(y_test)
```

### ***ONE HOT ENCODING IN TOO MANY CATEGORIES***

```python
counts=dff['brand'].value_counts()
```

```python
dff['brand'].nunique()
threshold= 100 #this is like the values which are less than 100 will be converted to others
```

```python
replace=counts[counts <= threshold].index  #this tells us which are the values less than 100
```

```python
pd.get_dummies(dff['brand'].replace(replace,'others'))
```
