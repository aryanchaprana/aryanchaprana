### ***NORMALIZATION***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class lable','Alcohol','Mallic Acid']
```

```python
df.sample(6)
```

```python
X=df.iloc[:,1:]
y=df.iloc[:,0]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=43)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train)
X_trainn=scaler.transform(X_train)
X_testt=scaler.transform(X_test)
```

```python
X_trainn=pd.DataFrame(X_trainn,columns=X_train.columns)
X_testt=pd.DataFrame(X_testt,columns=X_test.columns)
```
