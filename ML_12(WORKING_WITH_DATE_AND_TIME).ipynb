converting object data type to datetime

```python
date['date']=pd.to_datetime(date['date'])
```

Extracting year

```python
date['year']=date['date'].dt.year
```

```python
date
```

Extracting months

```python
date['month']=date['date'].dt.month #agar month name chaiye to dt.month_name()
```

```python
date
```

Extracting day

```python
date['day']=date['date'].dt.day # if you want to know which day of the week use dt.dayofweek,aur agar name bhe chaiye to dt.day_name()
```

```python
date
```

if you want to know if the day is weekend

```python
date['day_name']=date['date'].dt.day_name()
```

```python
date
```

```python
date['day_is_weekend']=np.where(date['day_name'].isin(['Saturday','Sunday']),1,0)
```

```python
date
```

```python
date.drop(columns=['product_id','city_id','orders'])
```

Extracting week of the year

```python
date['week_of_year']=date['date'].dt.isocalendar().week #dt.quarter se quarter mil jata hai
```
